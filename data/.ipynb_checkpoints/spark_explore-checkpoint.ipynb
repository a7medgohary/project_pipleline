```python
from pyspark.sql import SparkSession

# إنشاء جلسة Spark
spark = SparkSession.builder.appName("TaxiDataExploration").getOrCreate()

# قراءة ملف parquet (المسار الصحيح داخل الحاوية)
df = spark.read.parquet("/home/jovyan/work/data/raw")

# عرض شكل البيانات
df.printSchema()
df.limit(10).toPandas()
```

```python
from pyspark.sql.functions import col, hour, dayofweek, count

# استخراج ساعة بداية الرحلة
df_with_hour = df.withColumn("pickup_hour", hour(col("tpep_pickup_datetime")))

# حساب عدد الرحلات لكل ساعة
hourly_counts = df_with_hour.groupBy("pickup_hour").count().orderBy("pickup_hour")
hourly_counts.show()
```
